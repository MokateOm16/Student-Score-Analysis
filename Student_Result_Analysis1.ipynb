# Drop Unname column

```python
df = df.drop("Unnamed: 0",axis=1)
df.head()
```

# change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(data = df,x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# from the above chart  we have analysed that:
#the number of female in the data is more than the number of male
```

```python
gp = df.groupby("ParentEduc").agg({"MathScore": 'mean',"ReadingScore": 'mean',"WritingScore": 'mean',})
print(gp)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gp, annot = True)
plt.title("Relation Between Parent's Education and Student's Score")
plt.show()
```

```python
# from the above chart we have cocluded that the education of the  parent have a good impact
```

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean',"ReadingScore": 'mean',"WritingScore": 'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb ,annot=True)
plt.title("Relationship between Parents Marital Status and Student's Score")
plt.show()
```

```python
# from the above chart we have concluded that there is no/negligible impact on the 
#student score to their marital status
```

# Distirbution of Ethinic Group

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(data = df,x = "EthnicGroup")
ax.bar_label(ax.containers[0])
plt.title("Distirbution of Ethinic Group")
plt.show()
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
print(groupA)
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

explode = (0, 0, 0.1, 0,0) 
l = ["group A","group B","group C","group D","group E"]
mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mylist, explode=explode , labels = l,autopct = "%1.2f%%",shadow=True, startangle=140)
plt.title("Distirbution of Ethinic Group")
plt.show()
```
